```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

```python
price=[100,90,80,50,30,89]
hourse_power=[120,118,90,89,76,100]
mpg=[31,33,32,66,77,34]
```

```python
price.sort()
hourse_power.sort()
mpg.sort()
```

```python
plt.title('Plotting the value ')
plt.plot(price,marker='.',color='r')
plt.plot(hourse_power,marker='*',color='aqua')
plt.plot(mpg,marker='+',color='g')
plt.grid()
plt.show()
```

# Scatter plot

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 10 values for age and weight
age = np.random.uniform(40, 60, 10)
weight = np.random.uniform(60, 90, 10)

# Plot the age and weight values
plt.title('Plot of Age and Weight')
plt.scatter(age, weight, marker='*', color='hotpink')
plt.xlabel('Age')
plt.ylabel('Weight')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, c=colors)

plt.show()
```

# Bar Charts

```python
x=np.array(['A','B','C','D'])
y=np.array([30,40,50,60])
plt.bar(x,y,color='hotpink')

plt.title('Bar Charts')
plt.grid()
plt.show()
```

# horizintol bar

```python
fruits=np.array(['mango','orange','banana','watermelon'])
price_kg=([100,120,140,80])

plt.barh(fruits,price_kg,color='aqua')
plt.grid()
plt.show()
```

# Pie cahrt

```python
y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.title('Pie Charts')
plt.grid()
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show()
```
